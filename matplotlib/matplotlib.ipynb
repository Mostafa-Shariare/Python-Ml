```python
# importing
import matplotlib.pyplot as plt
```

```python
#import numpy to get data fpr out plots
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

plotting the data

```python
# sine wave
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, z)
plt.show()
```

```python
# adding title to x-axsis and y-axsis labels
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Wave')
plt.show()
```

```python
# plot a parabola
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Parabola')
plt.show()
```

```python
plt.plot(x, y, 'r+')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Parabola')
plt.show()
```

```python
plt.plot(x, y, 'g--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Parabola')
plt.show()
```

```python
plt.plot(x, y, 'rx')
plt.show()
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g--')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

```python
# bar plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'French', 'German', 'Bangla', 'Chinese']
people = [100, 85, 67, 43, 21]
ax.bar(languages, people)
plt.xlabel('Languages')
plt.ylabel('Number of People')
plt.title('Programming Languages')
plt.show()
```

```python
#Pie Chart
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])
languages = ['English', 'French', 'German', 'Bangla', 'Chinese']
people = [100, 85, 67, 43, 21]
ax1.pie(people, labels = languages, autopct='%1.1f')
plt.show()
```

```python
#Scatter Plot
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax2 = fig2.add_axes([0,0,1,1])
ax2.scatter(x,y,color='r')
ax2.scatter(x,z,color='b')
plt.show()
```

```python
#3d scatter plot
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c=z, cmap='Greens')
plt.show()
```
